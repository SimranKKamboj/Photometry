# Scaling the stars mag values

```python
import pandas as pd

data_7 = pd.read_csv('star12_24.csv')
data_8 = pd.read_csv('star2_24.csv')
data_9 = pd.read_csv('star12_29.csv')

scaler_mag7 = data_7['MAG7'] - data_8['MAG7']
scaler_mag5 = data_7['MAG5'] - data_8['MAG5']

scaler_merr7 = data_7['MAGUNCERT7'] - data_8['MAGUNCERT7']
scaler_merr5 = data_7['MAGUNCERT5'] - data_8['MAGUNCERT5']

final_mag7 = data_9['MAG7'] - scaler_mag7
final_mag5 = data_9['MAG5'] - scaler_mag5
final_merr7 = data_9['MAGUNCERT7'] - scaler_merr7
final_merr5 = data_9['MAGUNCERT5'] - scaler_merr5

# Creating a new DataFrame for final_mag and final_merr
final_data = pd.DataFrame({'final_mag7': final_mag7,
                           'final_mag5': final_mag5,
                           'final_merr7': final_merr7,
                           'final_merr5': final_merr5})

# Saving to CSV
final_data.to_csv('star12_29cor.csv', index=False)
```
